# Surface Plots 3D

```python
# Import modules
import matplotlib.pyplot as plt        # Graphics
import numpy as np                     # Math things
from mpl_toolkits import mplot3d       # 3D graphics tools
```

```python
def z_func(x, y):
    return (1 - (x ** 2 + y ** 3)) * np.exp(-(x ** 2 + y ** 2) / 2)

# Setting up input values
x = np.arange(-3.0, 3.0, 0.1)
y = np.arange(-3.0, 3.0, 0.1)
X, Y = np.meshgrid(x, y)

# Calculating the output and storing it in the array Z
Z = z_func(X, Y)

%matplotlib inline
fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 100, cmap='Greens')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
# Cambio del punto de vista
ax.view_init(75, 45)
fig
```

```python
# Podemos ver la superficie como un modelo alambre
fig = plt.figure(figsize=(8,6))
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color='red')
ax.set_title('wireframe')
ax.view_init(75, 45)
```

```python
# Se pueden agregar polígonos para ver la superficie de manera sólida
fig = plt.figure(figsize=(8,6))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
cmap='coolwarm', edgecolor='none')

ax.set_title('surface')
ax.view_init(75, 45)
```

```python
ax = plt.axes(projection="3d")                # Proyectaremos una gráfica

x_data = np.arange(-5, 5, 0.1)                # Datos
y_data = np.arange(-5, 5, 0.1)

X, Y = np.meshgrid(x_data, y_data)            # Malla del dominio
Z = np.sin(X) * np.cos(Y)

ax.plot_surface(X, Y, Z, cmap="plasma")
ax.view_init(azim=0, elev=90)
plt.show()
```

```python
ax = plt.axes(projection="3d")

def z_function(x,y):
    return np.sin(np.sqrt(x**2 + y**2))

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x, y)
Z = z_function(X, Y)

ax.plot_surface(X, Y, Z)
plt.show()
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D  
# Axes3D import has side effects, it enables using projection='3d' in add_subplot
import matplotlib.pyplot as plt
import random

# Define a function to pot it 
def fun(x,y):
    return x**2 + y

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = y = np.arange(-3.0, 3.0, 0.05)
X, Y = np.meshgrid(x, y)
zs = np.array(fun(np.ravel(X), np.ravel(Y)))
Z = zs.reshape(X.shape)

ax.plot_surface(X, Y, Z)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
