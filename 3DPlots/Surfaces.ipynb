```python
# Import modules
%matplotlib inline
import matplotlib.pyplot as plt        # Graphics
import numpy as np                     # Math things
from mpl_toolkits import mplot3d       # 3D graphics tools
```

```python
# Function
def func(x, y):
    if x>0:
        return np.sin(x) * np.cos(y)
    else:
        return np.sin(x-y)

# Now the function accepts vectors
func = np.vectorize(func)  
```

```python
# Setting up input values
x_data = np.arange(-6, 6, 0.1)                
y_data = np.arange(-6, 6, 0.1)
X, Y = np.meshgrid(x_data, y_data)

# Setting up output values
Z = func(X, Y)

def Animation(theta=-45):
    # Plotting
    fig = plt.figure(figsize=(12,10))
    ax = plt.axes(projection='3d') # projection

    surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='ocean') # settings
    ax.view_init(azim=theta, elev=40) # view

    plt.colorbar(surf, shrink=0.3, aspect=7)

    plt.title(r'$\sin(x)*\cos(x)$')
    ax.set_xlabel('x', labelpad=20)
    ax.set_ylabel('y', labelpad=20)
    ax.set_zlabel('u', labelpad=2)

    #ax.set_xlim(-6, 6)
    #ax.set_ylim(-6, 6)
    ax.set_zlim(-2, 2)

    #plt.savefig(f'picture_{theta}.png')
    plt.show()
```

```python
Animation(-45)
```

```python
# Generarmos 100 imgenes en funcin del angulo theta
rs = np.linspace(0, 180, 100)
for r in rs:
    Animation(r)
```

```python
# Importamos el modulo
import imageio

# Creamos una lista con el nombre de las imagenes
file_nms = [f'picture_{r}.png' for r in rs]

# Creamos una lista de imagenes con el mtodo imread
images = []
for file_name in file_nms:
    images.append(imageio.imread(file_name))

# Generamos el gif
imageio.mimsave('SurfaceMovie.gif', images)
```
